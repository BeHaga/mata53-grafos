```python
from pathlib import Path
import networkx as nx
from tqdm import tqdm
```

```python
communities = []

with open("../data/raw/com-orkut.top5000.cmty.txt", "r") as f:
    for line in f:
        nodes = line.strip().split()
        communities.append(nodes)

largest_community = max(
    communities,
    key=len
)

print("Tamanho:", len(largest_community))
```

```python
community_nodes = set(
    map(int, largest_community)
)

print("Nós da comunidade:", len(community_nodes))
```

```python
subgraph_edges = []

with open("../data/raw/com-orkut.ungraph.txt", "r") as f:

    for line in tqdm(f):

        if line.startswith("#"):
            continue

        u, v = map(int, line.split())

        if (
            u in community_nodes and
            v in community_nodes
        ):
            subgraph_edges.append((u, v))

print("Arestas encontradas:", len(subgraph_edges))
```

```python
G = nx.Graph()

G.add_edges_from(subgraph_edges)

print("Nós:", G.number_of_nodes())
print("Arestas:", G.number_of_edges())
```

```python
Path("../data/processed").mkdir(
    parents=True,
    exist_ok=True
)

nx.write_edgelist(
    G,
    "../data/processed/orkut_largest_community.edgelist",
    data=False
)
```
