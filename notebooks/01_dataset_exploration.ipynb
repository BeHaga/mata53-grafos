```python
community_sizes = []

with open("../data/raw/com-orkut.top5000.cmty.txt", "r") as f:
    for line in f:
        nodes = line.strip().split()
        community_sizes.append(len(nodes))

print("Número de comunidades:", len(community_sizes))

print("Maior comunidade:", max(community_sizes))

print("Top 10 tamanhos:")
print(sorted(community_sizes, reverse=True)[:10])
```
